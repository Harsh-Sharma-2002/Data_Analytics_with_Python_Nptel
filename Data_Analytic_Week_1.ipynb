#### **Min and Max Values**

```python
def min_max(col):
  min_val = min(col)
  max_val = max(col)
  return min_val,max_val
```

```python
min_total,max_total = min_max(df_2['Total'])
print('minimum :',min_total)
print('maximum :',max_total)
```

```python
def range_col(col):
  min_val = min(col)
  max_val = max(col)
  return (max_val - min_val)
```

```python
range = range_col(df_2['Total'])
print('range of data :',range)
```

#### **Quartile**

```python
q1 = np.percentile(df_2['Total'],25)
q2 = np.percentile(df_2['Total'],50) # Median
q3 = np.percentile(df_2['Total'],75)
q1,q2,q3
```

```python
# Inter quartile range
q3 - q1
```
